```python
import numpy as np
import matplotlib.pyplot as plt
from math import erf, sqrt

plt.rcParams.update({"font.size": 11})

c = 8
n = 25
eta_disc = np.linspace(0, 1, n + 1)
x_disc = []
for i in range(0, len(eta_disc)):
    x_disc.append(1 - (erf(sqrt(c)*(1-eta_disc[i])))/erf(sqrt(c)))
    
eta_cont = np.linspace(0, 1, 5*n)
x_cont = []
for i in range(0, len(eta_cont)):
    x_cont.append(1 - (erf(sqrt(c)*(1-eta_cont[i])))/erf(sqrt(c)))
    

plt.plot(eta_cont, x_cont)    
plt.scatter(eta_disc, x_disc, color = "red", s = 15)
plt.xlabel("$\eta$", fontsize = 14)
plt.ylabel("$x$", fontsize = 14)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from math import erf, sqrt, sinh, tanh

plt.rcParams.update({"font.size": 11})

#comparing each of the three mappings when all their constants a,b,c are 8
c = 8
n = 25
    
eta = np.linspace(0, 1, 5*n)

x_erf = []
for i in range(0, len(eta)):
    x_erf.append(1 - (erf(sqrt(c)*(1-eta[i])))/erf(sqrt(c)))
    
x_sinh = []
for i in range(0, len(eta)):
    x_sinh.append(sinh(c*eta[i])/sinh(c))
    
x_tanh = []
for i in range(0, len(eta)):
    x_tanh.append(1 - (tanh(c*(1-eta[i])))/tanh(c))

plt.plot(eta, x_erf, color = "red", label="erf")    
plt.plot(eta, x_sinh, color = "blue", label="sinh") 
plt.plot(eta, x_tanh, color = "black", label="tanh") 
plt.legend(loc="upper left")
plt.xlabel("$\eta$", fontsize = 16)
plt.ylabel("$x$", fontsize = 16)
plt.show()
```
